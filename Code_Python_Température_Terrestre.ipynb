```python
#Code Graphique 1 : Evolution annuelle de la température mondiale (moyenne annuelle)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Charger les données depuis l'URL
glob_url = "https://data.giss.nasa.gov/gistemp/tabledata_v4/GLB.Ts+dSST.csv"
df = pd.read_csv(glob_url, skiprows=1)  # Ignore la première ligne inutile

# Remplacer les valeurs "***" par des chaînes vides ""
df.replace("***", "", inplace=True)

# Transformer les valeurs '-.37' en '-0.37'
df = df.applymap(lambda x: str(x).replace('-.', '-0.') if isinstance(x, str) else x)

# Convertir toutes les colonnes (sauf "Year") en numériques
cols_to_convert = df.columns.difference(["Year"])  # Exclure "Year"
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# Convertir "Year" en entier
df["Year"] = pd.to_numeric(df["Year"], errors='coerce').astype("Int64")

# Visualisation
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='J-D', color='red')
plt.title("Evolution de la température Mondiale par année (Moyenne annuelle)")
plt.xlabel('Année')
plt.ylabel('Température (°C)')
plt.grid(True)
plt.show()
```

```python
#Code Graphique 2 : Températures mondiales par saison et par année (moyennes saisonnières)

plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['DJF'], label='Hiver (DJF)', marker='o', linestyle='-')
plt.plot(df['Year'], df['MAM'], label='Printemps (MAM)', marker='o', linestyle='-')
plt.plot(df['Year'], df['JJA'], label='Été (JJA)', marker='o', linestyle='-')
plt.plot(df['Year'], df['SON'], label='Automne (SON)', marker='o', linestyle='-')

plt.title('Evolution de la température Mondiale par saisons et par année (Moyennes saisonnières)')
plt.xlabel('Année')
plt.ylabel('Température (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Code Graphique 8 : Evolution de la température par zone de latitude

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Définir l'URL des données zonales
zonal_url = 'https://data.giss.nasa.gov/gistemp/tabledata_v4/ZonAnn.Ts+dSST.csv'

# Spécifier les noms des colonnes manuellement
columns = ['Year', 'Glob', 'NHem', 'SHem', '24N-90N', '24S-24N', '90S-24S',
           '64N-90N', '44N-64N', '24N-44N', 'EQU-24N', '24S-EQU',
           '44S-24S', '64S-44S', '90S-64S', 'Year_duplicate']

# Charger les données dans un DataFrame pandas avec les colonnes spécifiées
zonal_data = pd.read_csv(zonal_url, names=columns, skiprows=1)

# Supprimer la colonne Year_duplicate, qui est une répétition de Year
zonal_data.drop('Year_duplicate', axis=1, inplace=True)

# Bandes de latitude
bandes = ['64N-90N', '44N-64N', '24N-44N', 'EQU-24N', '24S-EQU', '44S-24S', '64S-44S', '90S-64S']

# Créer des subplots pour chaque bande de latitude
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(16, 16), sharex=True, sharey=True)
axes = axes.flatten()

for i, bande in enumerate(bandes):
    axes[i].plot(zonal_data['Year'], zonal_data[bande], label=f'Température {bande}')
    axes[i].set_title(f'Tendance des Températures pour la Bande {bande}')
    axes[i].set_xlabel('Année')
    axes[i].set_ylabel('Température Annuelle Moyenne Anomalie (°C)')
    axes[i].legend()
    axes[i].grid(True)

# Ajuster l'espacement entre les subplots
plt.tight_layout()
plt.show()
```

```python
#Rapport de Modélisation 1
#Code Graphique 1 : Température mondiale par année Régression Ridge

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV
import seaborn as sns

data_by_country  = pd.read_csv('https://data.giss.nasa.gov/gistemp/tabledata_v3/GLB.Ts+dSST.csv', skiprows=1)
temperature_data = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')

# Remplace valeurs manquantes par NaN et suppression lignes
data_by_country  = data_by_country .replace('***', np.nan).dropna()
# Conversion de la colonne 'J-D' en type floating pour la modélisation
data_by_country ['J-D'] = data_by_country ['J-D'].astype(float)

# Sélection de colonnes spécifiques et Suppression des lignes où co2 est manquant (NaN)
colonnes = ['year', 'gdp', 'co2', 'population', 'cement_co2', 'coal_co2', 'oil_co2', 'gas_co2', 'methane', 'nitrous_oxide']
temperature_data = temperature_data[colonnes].dropna(subset=['co2'])
# Regroupement par année (year) et Calcul de la moyenne pour chaque année
co2_moy = temperature_data.groupby('year').mean().reset_index()

# Fusion des data Temp avec data CO2 agrégées sur l'année
df = pd.merge(data_by_country, co2_moy, left_on='Year', right_on='year').dropna() # fusion avec année comme clé de jointure pour avoir une ligne de données complète par année

# Variables explicatives (X) et variable cible (y)
X = df[['year', 'gdp', 'co2', 'population', 'cement_co2', 'coal_co2', 'oil_co2', 'gas_co2', 'methane', 'nitrous_oxide']]

# Variable cible anomalie moyenne de température 'J-D'
y = df['J-D']

# StandardScaler normalise les données explicatives (moyenne de 0 et une variance de 1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Entraînement (80%) et test (20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Sélection de l'hyperparamètre alpha avec validation croisée
ridge = RidgeCV(alphas=[0.1, 1.0, 10.0, 100.0], store_cv_values=True)
ridge.fit(X_train, y_train)

# Prédictions
y_test_pred = ridge.predict(X_test)

# Évaluation du modèle
print(f"Meilleur alpha: {ridge.alpha_}")
print("Test R²:", r2_score(y_test, y_test_pred))
print("Train R²:", r2_score(y_train, ridge.predict(X_train)))
print("Test MAE:", mean_absolute_error(y_test, y_test_pred))
print("Test RMSE:", np.sqrt(mean_squared_error(y_test, y_test_pred)))

# Visualisation des prédictions vs les vraies valeurs
plt.scatter(y_test, y_test_pred, alpha=0.6, color='b')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='r', linestyle='--')
plt.xlabel('Valeurs réelles (y_test)')
plt.ylabel('Prédictions (y_test_pred)')
plt.title('Prédictions vs Réalité')
plt.show()
```

```python
#Code Complet pour Rapport de Modélisation 3


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.neural_network import MLPRegressor

# Chargement des données de température (GISS NASA)
temp_data = pd.read_csv('https://data.giss.nasa.gov/gistemp/tabledata_v3/GLB.Ts+dSST.csv', skiprows=1)

# Chargement des données de CO2 (OWID)
co2_data = pd.read_csv('https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv')

# Nettoyage des données de température
temp_data = temp_data.replace('***', np.nan).dropna()
temp_data['J-D'] = temp_data['J-D'].astype(float)

# Sélection des colonnes pertinentes dans les données de CO2
co2_cols = ['year', 'gdp', 'co2', 'population', 'cement_co2', 'coal_co2', 'oil_co2', 'gas_co2', 'methane', 'nitrous_oxide']
co2_data = co2_data[co2_cols].dropna(subset=['co2'])

# Agrégation des données de CO2 par année
co2_agg = co2_data.groupby('year').mean().reset_index()

# Fusion des données de température et de CO2 sur l'année commune
df = pd.merge(temp_data, co2_agg, left_on='Year', right_on='year').dropna()

# Définition des variables explicatives et de la variable cible
X = df[['year', 'gdp', 'co2', 'population', 'cement_co2', 'coal_co2', 'oil_co2', 'gas_co2', 'methane', 'nitrous_oxide']]
y = df['J-D']

# Normalisation des variables explicatives
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Séparation des données en ensembles d'entraînement et de test (70% / 30%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Entraînement du modèle de Régression Linéaire
model_lr = LinearRegression().fit(X_train, y_train)

# Prédictions sur les données de test
y_test_pred_lr = model_lr.predict(X_test)

# Évaluation du modèle de Régression Linéaire
print("Validation Régression Linéaire:")
print(f"R² = {r2_score(y_test, y_test_pred_lr)}")
print(f"MAE = {mean_absolute_error(y_test, y_test_pred_lr)}")
print(f"RMSE = {np.sqrt(mean_squared_error(y_test, y_test_pred_lr))}")
print(f"MSE = {mean_squared_error(y_test, y_test_pred_lr)}")

# Validation croisée du modèle de Régression Linéaire
scores_lr = cross_val_score(model_lr, X_train, y_train, cv=5, scoring='r2')
print(f"Cross-Validation Scores - Régression Linéaire: {scores_lr.mean()}")

# Visualisation des prédictions de validation croisée pour la Régression Linéaire
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_train, y=model_lr.predict(X_train), color='blue', alpha=0.6, label='Prédictions (Régression Linéaire)')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], color='red', linestyle='-', linewidth=2)
plt.xlabel('Valeurs réelles (y_train)')
plt.ylabel('Prédictions (y_train_pred)')
plt.title('Validation Croisée - Prédictions vs Valeurs Réelles - Régression Linéaire')
plt.legend()
plt.show()

# Définir la grille de paramètres pour le GridSearchCV
param_grid_rf = {'n_estimators': [100, 200, 500], 'max_depth': [10, 20, 30]}
grid_search_rf = GridSearchCV(estimator=RandomForestRegressor(random_state=42), param_grid=param_grid_rf, cv=5, scoring='r2')

# Effectuer la recherche sur grille avec les données d'entraînement
grid_search_rf.fit(X_train, y_train)

# Extraction des résultats du GridSearchCV
results = pd.DataFrame(grid_search_rf.cv_results_)
best_params = grid_search_rf.best_params_

print("Meilleurs hyperparamètres trouvés :")
print(best_params)

print("Toutes les combinaisons d'hyperparamètres testées et les scores correspondants :")
print(results[['param_n_estimators', 'param_max_depth', 'mean_test_score', 'std_test_score']])

top_n = 5
top_results = results.nlargest(top_n, 'mean_test_score')
print(f"Top {top_n} combinaisons d'hyperparamètres :")
print(top_results[['param_n_estimators', 'param_max_depth', 'mean_test_score', 'std_test_score']])

# Utiliser les meilleurs hyperparamètres trouvés par le GridSearchCV
best_rf = RandomForestRegressor(n_estimators=best_params['n_estimators'], max_depth=best_params['max_depth'], random_state=42)
best_rf.fit(X_train, y_train)

# Prédictions sur les données de test
y_test_pred_rf = best_rf.predict(X_test)

# Évaluation du modèle de Forêts Aléatoires avec les meilleurs hyperparamètres
print("Validation Forêts Aléatoires avec les meilleurs hyperparamètres:")
print(f"R² = {r2_score(y_test, y_test_pred_rf)}")
print(f"MAE = {mean_absolute_error(y_test, y_test_pred_rf)}")
print(f"RMSE = {np.sqrt(mean_squared_error(y_test, y_test_pred_rf))}")
print(f"MSE = {mean_squared_error(y_test, y_test_pred_rf)}")

# Validation croisée du modèle de Forêts Aléatoires avec les meilleurs hyperparamètres
scores_rf = cross_val_score(best_rf, X_train, y_train, cv=5, scoring='r2')
print(f"Scores de validation croisée - Forêts Aléatoires avec les meilleurs hyperparamètres : {scores_rf.mean()}")

# Visualisation des prédictions de validation croisée pour les Forêts Aléatoires
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_train, y=best_rf.predict(X_train), color='green', alpha=0.6, label='Prédictions (Forêts Aléatoires)')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], color='red', linestyle='-', linewidth=2)
plt.xlabel('Valeurs réelles (y_train)')
plt.ylabel('Prédictions (y_train_pred)')
plt.title('Validation Croisée - Prédictions vs Valeurs Réelles - Forêts Aléatoires')
plt.legend()
plt.show()

# Régression Ridge
degree = 2
ridge_model = make_pipeline(PolynomialFeatures(degree), Ridge(alpha=1.0))
ridge_model.fit(X_train, y_train)
y_test_pred_ridge = ridge_model.predict(X_test)

# Évaluation du modèle de régression Ridge
print("Validation Régression Ridge:")
print(f"R² = {r2_score(y_test, y_test_pred_ridge)}")
print(f"MAE = {mean_absolute_error(y_test, y_test_pred_ridge)}")
print(f"RMSE = {np.sqrt(mean_squared_error(y_test, y_test_pred_ridge))}")
print(f"MSE = {mean_squared_error(y_test, y_test_pred_ridge)}")

# Visualisation des prédictions de validation croisée pour la Régression Ridge
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_train, y=ridge_model.predict(X_train), color='orange', alpha=0.6, label='Prédictions (Ridge)')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], color='red', linestyle='-', linewidth=2)
plt.xlabel('Valeurs réelles (y_train)')
plt.ylabel('Prédictions (y_train_pred)')
plt.title('Validation Croisée - Prédictions vs Valeurs Réelles - Régression Ridge')
plt.legend()
plt.show()

# Gradient Boosting
gb_model = GradientBoostingRegressor()
gb_model.fit(X_train, y_train)
y_test_pred_gb = gb_model.predict(X_test)

# Évaluation du modèle de Gradient Boosting
print("Validation Gradient Boosting:")
print(f"R² = {r2_score(y_test, y_test_pred_gb)}")
print(f"MAE = {mean_absolute_error(y_test, y_test_pred_gb)}")
print(f"RMSE = {np.sqrt(mean_squared_error(y_test, y_test_pred_gb))}")
print(f"MSE = {mean_squared_error(y_test, y_test_pred_gb)}")

# Visualisation des prédictions de validation croisée pour le Gradient Boosting
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_train, y=gb_model.predict(X_train), color='purple', alpha=0.6, label='Prédictions (GB)')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], color='red', linestyle='-', linewidth=2)
plt.xlabel('Valeurs réelles (y_train)')
plt.ylabel('Prédictions (y_train_pred)')
plt.title('Validation Croisée - Prédictions vs Valeurs Réelles - Gradient Boosting')
plt.legend()
plt.show()

# Réseau de Neurones
nn_model = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=1000)
nn_model.fit(X_train, y_train)
y_test_pred_nn = nn_model.predict(X_test)

# Évaluation du modèle de Réseau de Neurones
print("Validation Réseau de Neurones:")
print(f"R² = {r2_score(y_test, y_test_pred_nn)}")
print(f"MAE = {mean_absolute_error(y_test, y_test_pred_nn)}")
print(f"RMSE = {np.sqrt(mean_squared_error(y_test, y_test_pred_nn))}")
print(f"MSE = {mean_squared_error(y_test, y_test_pred_nn)}")

# Visualisation des prédictions de validation croisée pour le Réseau de Neurones
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y_train, y=nn_model.predict(X_train), color='green', alpha=0.6, label='Prédictions (NN)')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], color='red', linestyle='-', linewidth=2)
plt.xlabel('Valeurs réelles (y_train)')
plt.ylabel('Prédictions (y_train_pred)')
plt.title('Validation Croisée - Prédictions vs Valeurs Réelles - Réseau de Neurones')
plt.legend()
plt.show()

# Comparaison des scores pour chaque modèle
models = {
    'Régression Linéaire': model_lr,
    'Forêts Aléatoires': best_rf,
    'Régression Ridge': ridge_model,
    'Gradient Boosting': gb_model,
    'Réseau de Neurones': nn_model
}

for name, model in models.items():
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    print(f"{name} - Score d'entraînement: {train_score:.4f}")
    print(f"{name} - Score de test: {test_score:.4f}")

# Supposons que X_train, y_train, model_lr, best_rf, ridge_model, gb_model, nn_model soient déjà définis et entraînés

# Trouver la méthode avec les meilleurs résultats (ici, nous considérons le R² le plus élevé)
best_model_name = max(models, key=lambda name: r2_score(y_test, models[name].predict(X_test)))
best_model = models[best_model_name]

# Créer le graphique
plt.figure(figsize=(14, 10))

for model_name, model in models.items():
    sns.scatterplot(x=y_train, y=model.predict(X_train), label=model_name, alpha=0.6)

# Tracer la ligne pour le modèle avec les meilleurs résultats
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], color='red', linestyle='-', linewidth=2, label='Valeurs Parfaites')

plt.xlabel('Valeurs réelles (y_train)')
plt.ylabel('Prédictions (y_train_pred)')
plt.title('Validation Croisée - Prédictions vs Valeurs Réelles')
plt.legend()
plt.grid(True)

# Ajouter la ligne pour les meilleures prédictions
sns.lineplot(x=y_train, y=best_model.predict(X_train), color='black', linewidth=2, label=f'Meilleur Modèle ({best_model_name})')

plt.legend()
plt.show()

# Définir les années futures à prédire de 2025 à 2100
future_years = np.arange(2025, 2100)

growth_rates_optimistic = {
    'gdp': 0.025,  # Ajustement à une croissance modérée (2-3 %)
    'co2': -0.02,  # Taux de croissance négatif pour refléter une réduction des émissions
    'population': 0.005,  # Croissance faible (0,5 %)
    'cement_co2': 0.005,  # Faible croissance avec technologies de capture du carbone
    'coal_co2': -0.02,  # Taux de croissance négatif pour réduire l'utilisation des énergies fossiles
    'oil_co2': -0.02,  # Taux de croissance négatif pour réduire l'utilisation des énergies fossiles
    'gas_co2': -0.02,  # Taux de croissance négatif pour réduire l'utilisation des énergies fossiles
    'methane': 0.005,  # Faible croissance (0,5 %)
    'nitrous_oxide': 0.005  # Faible croissance (0,5 %)
}


def adjusted_variation(mean_value, annual_growth_rate, length):
    return mean_value * (1 + annual_growth_rate) ** np.arange(length)

# Générer des données futures pour le scénario optimiste
future_data_optimistic = pd.DataFrame({
    'year': future_years,
    'gdp': adjusted_variation(df['gdp'].mean(), growth_rates_optimistic['gdp'], len(future_years)),
    'co2': adjusted_variation(df['co2'].mean(), growth_rates_optimistic['co2'], len(future_years)),
    'population': adjusted_variation(df['population'].mean(), growth_rates_optimistic['population'], len(future_years)),
    'cement_co2': adjusted_variation(df['cement_co2'].mean(), growth_rates_optimistic['cement_co2'], len(future_years)),
    'coal_co2': adjusted_variation(df['coal_co2'].mean(), growth_rates_optimistic['coal_co2'], len(future_years)),
    'oil_co2': adjusted_variation(df['oil_co2'].mean(), growth_rates_optimistic['oil_co2'], len(future_years)),
    'gas_co2': adjusted_variation(df['gas_co2'].mean(), growth_rates_optimistic['gas_co2'], len(future_years)),
    'methane': adjusted_variation(df['methane'].mean(), growth_rates_optimistic['methane'], len(future_years)),
    'nitrous_oxide': adjusted_variation(df['nitrous_oxide'].mean(), growth_rates_optimistic['nitrous_oxide'], len(future_years))
}, index=future_years)

# Normaliser les données futures
future_data_scaled_optimistic = scaler.transform(future_data_optimistic)

# Prédictions futures avec les meilleurs modèles
future_pred_lr_optimistic = model_lr.predict(future_data_scaled_optimistic)
future_pred_ridge_optimistic = ridge_model.predict(future_data_scaled_optimistic)
future_pred_nn_optimistic = nn_model.predict(future_data_scaled_optimistic)

# Combiner les prédictions futures (moyenne simple)
future_ensemble_pred_optimistic = (future_pred_lr_optimistic + future_pred_ridge_optimistic + future_pred_nn_optimistic) / 3

# Créer un DataFrame avec les prédictions futures pour le scénario optimiste
future_predictions_df_optimistic = pd.DataFrame({
    'Year': future_years,
    'Predicted_Temperature_Anomaly': future_ensemble_pred_optimistic
})

# Afficher le DataFrame pour le scénario optimiste
print("Prédictions futures - Scénario Optimiste :")
print(future_predictions_df_optimistic)

# Créer un graphique des prédictions futures de 2025 à 2100 pour le scénario optimiste
plt.figure(figsize=(14, 7))
plt.plot(future_predictions_df_optimistic['Year'], future_predictions_df_optimistic['Predicted_Temperature_Anomaly'], marker='o', linestyle='-', color='b', label='Scénario Optimiste')
plt.axhline(y=1.5, color='r', linestyle='--', label='Limite de 1,5°C (Accord de Paris)')
plt.xlabel('Année')
plt.ylabel('Anomalie de Température Prédite (°C)')
plt.title('Prédictions Futures des Anomalies de Température (2025-2100) - Scénario Optimiste')
plt.legend()
plt.grid(True)
plt.show()

growth_rates_pessimistic = {
    'gdp': 0.03,  # Croissance modérée (3 %)
    'co2': 0.025,  # Taux de croissance élevé mais réaliste pour les émissions de CO2
    'population': 0.01,  # Croissance modérée (1 %)
    'cement_co2': 0.02,  # Taux de croissance modéré pour le ciment
    'coal_co2': 0.015,  # Taux de croissance élevé pour le charbon
    'oil_co2': 0.015,  # Taux de croissance élevé pour le pétrole
    'gas_co2': 0.015,  # Taux de croissance élevé pour le gaz
    'methane': 0.015,  # Croissance modérée pour le méthane
    'nitrous_oxide': 0.015  # Croissance modérée pour l'oxyde nitreux
}

# Générer des données futures pour le scénario pessimiste
future_data_pessimistic = pd.DataFrame({
    'year': future_years,
    'gdp': adjusted_variation(df['gdp'].mean(), growth_rates_pessimistic['gdp'], len(future_years)),
    'co2': adjusted_variation(df['co2'].mean(), growth_rates_pessimistic['co2'], len(future_years)),
    'population': adjusted_variation(df['population'].mean(), growth_rates_pessimistic['population'], len(future_years)),
    'cement_co2': adjusted_variation(df['cement_co2'].mean(), growth_rates_pessimistic['cement_co2'], len(future_years)),
    'coal_co2': adjusted_variation(df['coal_co2'].mean(), growth_rates_pessimistic['coal_co2'], len(future_years)),
    'oil_co2': adjusted_variation(df['oil_co2'].mean(), growth_rates_pessimistic['oil_co2'], len(future_years)),
    'gas_co2': adjusted_variation(df['gas_co2'].mean(), growth_rates_pessimistic['gas_co2'], len(future_years)),
    'methane': adjusted_variation(df['methane'].mean(), growth_rates_pessimistic['methane'], len(future_years)),
    'nitrous_oxide': adjusted_variation(df['nitrous_oxide'].mean(), growth_rates_pessimistic['nitrous_oxide'], len(future_years))
}, index=future_years)

# Normaliser les données futures
future_data_scaled_pessimistic = scaler.transform(future_data_pessimistic)

# Prédictions futures avec les meilleurs modèles
future_pred_lr_pessimistic = model_lr.predict(future_data_scaled_pessimistic)
future_pred_ridge_pessimistic = ridge_model.predict(future_data_scaled_pessimistic)
future_pred_nn_pessimistic = nn_model.predict(future_data_scaled_pessimistic)

# Combiner les prédictions futures (moyenne simple)
future_ensemble_pred_pessimistic = (future_pred_lr_pessimistic + future_pred_ridge_pessimistic + future_pred_nn_pessimistic) / 3

# Créer un DataFrame avec les prédictions futures pour le scénario pessimiste
future_predictions_df_pessimistic = pd.DataFrame({
    'Year': future_years,
    'Predicted_Temperature_Anomaly': future_ensemble_pred_pessimistic
})

# Afficher le DataFrame pour le scénario pessimiste
print("Prédictions futures - Scénario Pessimiste :")
print(future_predictions_df_pessimistic)

# Créer un graphique des prédictions futures de 2025 à 2100 pour le scénario pessimiste
plt.figure(figsize=(14, 7))
plt.plot(future_predictions_df_pessimistic['Year'], future_predictions_df_pessimistic['Predicted_Temperature_Anomaly'], marker='o', linestyle='-', color='r', label='Scénario Pessimiste')
plt.axhline(y=1.5, color='r', linestyle='--', label='Limite de 1,5°C (Accord de Paris)')
plt.xlabel('Année')
plt.ylabel('Anomalie de Température Prédite (°C)')
plt.title('Prédictions Futures des Anomalies de Température (2025-2100) - Scénario Pessimiste')
plt.legend()
plt.grid(True)
plt.show()

# Graphique de comparaison des predictions des deux différents scénarios OPTION
plt.figure(figsize=(14, 7))
plt.plot(future_predictions_df_optimistic['Year'], future_predictions_df_optimistic['Predicted_Temperature_Anomaly'], color='b', label='Scénario Optimiste')
plt.plot(future_predictions_df_pessimistic['Year'], future_predictions_df_pessimistic['Predicted_Temperature_Anomaly'], color='r', label='Scénario Pessimiste')
plt.axhline(y=1.5, color='g', linestyle='--', label='Limite de 1,5°C (Accord de Paris)')
plt.xlabel('Année')
plt.ylabel('Anomalie de Température Prédite (°C)')
plt.title('Comparaison des Scénarios Climatiques (2025-2100)')
plt.legend()
plt.grid(True)
plt.show()
```
